# RooFit flavour-physics hands-on

## 1. Setup the ROOT environment

```python
import subprocess
from pathlib import Path
import ROOT
from IPython.display import Image, display


ROOT.gROOT.SetBatch(True)

cwd = Path.cwd()
repo = cwd 
work = cwd / 'notebook_outputs'
plots = work / 'plots'

work.mkdir(exist_ok=True)
plots.mkdir(exist_ok=True)

def run_script(script, *args):
    command = ['python3', str(repo / 'scripts' / script), *map(str, args)]
    print(' '.join(command))
    subprocess.run(command, check=True)
```

## 2. Fit the signal mass and lifetime

The following example fits a signal only sample with a 2D model that parametrize the mass and decay time distribution. No background is ezpected.

Goal: review `scripts/tutorial_01_signal_fit.py`.
- How the data are prepared?
- How the probability density functions are defined?
- Are the fit results good? 
- What is the purpose of the 2D fit in this example? Is it needed?

```python
signal_fit_plot = plots / 'tutorial_01_signal_fit.png'
run_script('tutorial_01_signal_fit.py', '--input', 'data/signal_only.root', '--plot', signal_fit_plot)

display(Image(filename=str(signal_fit_plot)))
```

```python
mass_plot = plots / 'inspect_mass.png'
run_script(
    'inspect_branch.py',
    '--input', 'data/full.root',
    '--branch', 'mass',
    '--min', 5.0,
    '--max', 6.0,
    '--plot', mass_plot,
)
display(Image(filename=str(mass_plot)))

decay_time_plot = plots / 'inspect_decay_time.png'
run_script(
    'inspect_branch.py',
    '--input', 'data/full.root',
    '--branch', 'decay_time',
    '--min', 0.0,
    '--max', 13.0,
    '--plot', decay_time_plot,
)
display(Image(filename=str(decay_time_plot)))
```
